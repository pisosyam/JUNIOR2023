```python
import pandas
import numpy
import matplotlib.pyplot as plt

#Loading data
data = pandas.read_csv('data2_2.csv',header = None)
print (data.size)

#change our data
data = data.drop(data.index[[0,1,2]])
data = data.fillna(0.0)
data[0:] = data[0:].astype('float64')

#plotinng data
%matplotlib inline
plt.figure(figsize=(20,20))
i=1
while i <= 11:
    if i % 2 == 0:
        #plt.plot(data[[i]],data[[i-1]], linewidth=10-(i))
        plt.plot(data[[i]],data[[i-1]], linewidth=2 )
    i+=1
print ("done")
plt.xlim(0.5,3.0)
plt.legend(['1','2','3','4','5'])
#plt.savefig('methylimidazole_1_5.png')

plt.figure(figsize=(20,20))
i=11
while i <= 15:
    if i % 2 == 0:
        #plt.plot(data[[i]],data[[i-1]], linewidth=10-(i))
        plt.plot(data[[i]],data[[i-1]], linewidth=2 )
    i+=1
print ("done")
plt.xlim(0.5,3.0)
plt.legend(['6','7','8'])
#plt.savefig('methylimidazole_6_7.png')



plt.figure(figsize=(20,20))
i=15
while i <= 27:
    if i % 2 == 0:
        #plt.plot(data[[i]],data[[i-1]], linewidth=10-(i))
        plt.plot(data[[i]],data[[i-1]], linewidth=2 )
    i+=1
print ("done")
plt.xlim(0.5,3.0)
plt.legend(['8','9','10','11','12','13'])
#plt.savefig('methylimidazole_8_13.png')
```

```python
from matplotlib.pyplot import cm
#plotinng data
%matplotlib inline


plt.figure(figsize=(20,20))

color=iter(cm.rainbow(numpy.linspace(0,2,20)))
for i in range (1,11):
    if i % 2 == 0:
        c = next(color)
        plt.plot(data[[i]],data[[i-1]], color = c, linewidth=2 )
for i in range (11,17):
    print("nope")
for i in range (17,27):
    if i % 2 == 0:
        c=next(color)
        plt.plot(data[[i]],data[[i-1]],color =c, linewidth=2 )
print ("done")
plt.xlim(0.5,3.0)
plt.legend(['1','2','3','4','5','9','10','11','12','13'])
plt.savefig('methylimidazole_1_13.png')
```

```python
#Loading data
data = pandas.read_csv('data2_2.csv',header = None)
#print (data.size)

#change our data
data = data.drop(data.index[[0,1,2]])
data = data.fillna(0.0)
data[0:] = data[0:].astype('float64')



print ()
#x = max(data[[17]].idxmax())
#print (x)
for i in range(18,27):
    if i % 2 == 0:
        datamin =  data[[26]].loc[(max(data[[25]].idxmax()))+1].values - data[[8]].loc[(max(data[[7]].idxmax()))+1].values
        data[[i]] = data[[i]].subtract(datamin)

print ("end")
```
